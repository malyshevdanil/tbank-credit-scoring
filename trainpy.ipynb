```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import accuracy_score
import joblib
```

```python
df = pd.read_csv('scoring .csv')
df.head()
```

```python
X = df.drop(columns=['default']).values
y = df['default']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
X_train.shape
X_test.shape
```

## Tbank logisticRegression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(class_weight='balanced')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print(f"Отказано: {y_pred.mean() * 100:.0f}%")
print(f"Точность: {precision * 100:.0f}%")
print(f"Полнота: {recall * 100:.0f}%")
print(f"Accuracy: {accuracy_score(y_test, y_pred) * 100:.0f}%")
```

## Tbanl Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(class_weight='balanced', random_state=42)
model_rf.fit(X_train, y_train)
y_pred = model_rf.predict(X_test)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print(f"Отказано: {y_pred.mean() * 100:.0f}%")
print(f"Точность: {precision * 100:.0f}%")
print(f"Полнота: {recall * 100:.0f}%")
print(f"Accuracy: {accuracy_score(y_test, y_pred) * 100:.0f}%")
```

## Маштабирование Логистической регресии

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# Масштабирование
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Модель
model = LogisticRegression(class_weight='balanced')
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled) 

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print(f"Отказано: {y_pred.mean() * 100:.0f}%")
print(f"Точность: {precision * 100:.0f}%")
print(f"Полнота: {recall * 100:.0f}%")
print(f"Accuracy: {accuracy_score(y_test, y_pred) * 100:.0f}%")
joblib.dump(model, "model.pkl")
joblib.dump(scaler, "scaler.pkl")
```

```python
coefficients = pd.Series(model.coef_[0], index=['age', 'income', 'education', 'work', 'car'])
print(coefficients.sort_values())
```
